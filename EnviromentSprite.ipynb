```python
import pygame
import random
```

```python
#borders
WIDTH = 500
HEIGHT = 500
FPS =45
#colors
WHITE= (255,255,255)
BLACK= (0,0,0)
RED= (255,0,0)
GREEN= (0,255,0)
BLUE= (0,0,255)

max_speed=30
```

```python
class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)  
        self.image= pygame.Surface((25,25))
        self.image.fill(BLACK)
        self.rect= self.image.get_rect()
        self.radius=10
        pygame.draw.circle(self.image, RED, self.rect.center, self.radius)
        self.maxspeed = 4

        self.rect.x = random.randrange(0,WIDTH - self.rect.width) #Our enemy spawn inside of the borders, but it has 20,20 area so some pixels might spawn at outside of the borders. We fix that here
        self.rect.y = random.randrange(4,10) # spawn area at y 

        self.speedx= 0
        self.speedy= 6

    def update(self):
        self.rect.x += self.speedx
        self.rect.y += self.maxspeed

        if self.rect.top > HEIGHT + 20:
            self.rect.x = random.randrange(0,WIDTH - self.rect.width)
            self.rect.y = random.randrange(4,10)
            
            if self.maxspeed < 13:
                self.maxspeed +=1
                reward= self.maxspeed
                print(reward)

    def getCoordinates(self):
        return (self.rect.x, self.rect.y)
    
```

```python
pygame.init() # Initialize pygame 
screen = pygame.display.set_mode((WIDTH,HEIGHT)) #create window.
pygame.display.set_caption = ('RL GAME !')
clock = pygame.time.Clock()
```

```python
#sprite
all_sprite = pygame.sprite.Group()
player=Player()
enemy_group= pygame.sprite.Group()
m1= Enemy()
m2= Enemy()
m3= Enemy()
m4= Enemy()
all_sprite.add(player)
all_sprite.add(m1)
all_sprite.add(m2)
all_sprite.add(m3)
all_sprite.add(m4)
enemy_group.add(m1)
enemy_group.add(m2)
enemy_group.add(m3)
enemy_group.add(m4)
```

```python
# game loop
total_reward= 0
running = True

while running:
    clock.tick(FPS)

    # Process input
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Update
    player.update()
    enemy_group.update()
    all_sprite.update()

    hits = pygame.sprite.spritecollide(player, enemy_group, False, pygame.sprite.collide_circle)

    if hits:
        total_reward += sum([enemy.maxspeed for enemy in enemy_group])
        print("Total Points:", total_reward/4)
        running = False
        
    # Draw/render
    screen.fill(BLUE)
    all_sprite.draw(screen)
    pygame.display.flip()  # Göster

pygame.quit() 
```
